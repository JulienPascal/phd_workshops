### Exercise 1

Plot the tent map

$$ f(x) = \mu \min\{x, 1 - x\} $$

on the interval $[0, 1]$ when $\mu = 1.8$

```julia
for i in 1:15
    println("solution below")
end
```

```julia
μ = 1.8
x = linspace(0, 1, 200)
y = μ .* min.(x, 1 - x);
```

```julia
plot(x, y)
```

### Exercise 2

The following example makes an empty array of length 3 and then fills them with the numbers 0.5, 1.0, 0.5:

```julia
x = Array{Float64}(3)
x[1] = 0.5
x[2] = 1.0
x[3] = 0.5
```

Now compute the time series of length 500 given by

$$ x_{t+1} = f(x_t) $$

where $f$ is as given above and $x_0 = 0.2$.  Plot $x_t$ against $t$.

```julia
for i in 1:15
    println("solution below")
end
```

```julia
n = 500
x = Array{Float64}(n)
x[1] = 0.2
for t in 1:(n-1)
    x[t+1] = μ * min(x[t], 1 - x[t])
end
```

```julia
plot(x)
```
